```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_states = df.groupby('Ship-State')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_states.values, y=top_states.index,
            hue=top_states.index,
            palette='Blues_r',
           legend=False)

plt.title("Top 10 States by Sales")
plt.xlabel("Sales Amount")
plt.ylabel("States")
plt.show()
```

```python
top_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=top_category.values,
            y=top_category.index,
            hue=top_category.index,
            palette='viridis',
            legend=False)
plt.title("Sales by Category")
plt.xlabel("Sales Amount")
plt.ylabel("Category")

plt.show()
```

```python
status_count = df['Status'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=status_count.values,
    y=status_count.index,
    hue=status_count.index,
    palette='rocket',
    legend=False)

plt.title("Top Order Status")
plt.xlabel("Count")
plt.ylabel("Order Status")

plt.show()
```

```python
monthly_sales = df.groupby('Month')['Amount'].sum()

plt.figure(figsize=(8,5))

sns.lineplot(
    x=monthly_sales.index,
    y=monthly_sales.values,
    marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")

plt.show()
```
